# MURB Data Explorer

Quick-start notebook for querying and visualizing the MURB database.

```python
from murb_db import query, list_tables, describe_table, get_schema_summary, search_columns
from murb_db import bar_chart, scatter, timeseries
from murb_db.config import get_connection

conn = get_connection()
```

## 1. What tables do we have?

```python
tables = list_tables(conn=conn)
print(f"Tables ({len(tables)}):")
for t in tables:
    print(f"  - {t}")
```

## 2. Schema summary (AI-readable)

```python
print(get_schema_summary(conn=conn))
```

## 3. Preview a table

```python
# Replace 'your_table' with an actual table name from above
if tables:
    tbl = tables[0]
    print(f"Schema of '{tbl}':")
    display(describe_table(tbl, conn=conn))
    print(f"\nFirst 10 rows:")
    display(query(f"SELECT * FROM [{tbl}] LIMIT 10", conn=conn))
```

## 4. Search for columns

```python
# Find all columns containing 'energy'
search_columns("energy", conn=conn)
```
